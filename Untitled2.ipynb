```python
import pandas as pd
import matplotlib.pyplot as plt
import pandasql as sq 
import seaborn as sb
```

```python
data=pd.read_csv(r"C:/Users/ashishkumd2/Desktop/SOurcs/ML/Datsets/Automobile_data.csv")
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
data.head(100)
data.fillna(0,inplace=True)
data.replace("?",0,inplace=True)
cat=[]
con=[]
for i in data.columns:
    if data[i].dtypes=="object":
        data[i].replace(0,data[i].mode()[0],inplace=True)
        cat.append(i)
    else:
        data[i].replace(0,data[i].astype(int).mean(),inplace=True)
        con.append(i)
data['normalized-losses'].replace(0,data['normalized-losses'].astype(int).mean(),inplace=True)
data['price'].replace(0,data['price'].astype(int).mean(),inplace=True)
data
```

```python
for i in data.columns:
    if(data[i].dtypes=="object"):
        sb.boxplot(data[i],data.price.astype(int))
        plt.show()
    else:
        sb.scatterplot(data.price.astype(int),data[i])
        plt.show()
```

```python
data.columns
```

```python
X=data.drop(cat,axis=1)
Y=data['price'].astype(float)

from sklearn.preprocessing import StandardScaler
sl=StandardScaler()
X=sl.fit_transform(X)
X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestRegressor


from sklearn.metrics import accuracy_score,mean_absolute_error

xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.1,random_state=1)

sb.distplot(xtrain)
sb.distplot(xtest)
```

```python
#DecisionTree
dt=DecisionTreeRegressor(max_depth=4)
model=dt.fit(xtrain,ytrain)
pred=model.predict(xtest)
pd.DataFrame(zip(pred,ytest))
mean_absolute_error(ytest,pred)
```

```python
#KNN
kr=KNeighborsRegressor(n_neighbors=5)
model=kr.fit(xtrain,ytrain)
pred=model.predict(xtest)
pd.DataFrame(zip(pred,ytest))
mean_absolute_error(ytest,pred)
```

```python
#Logistic Regression
rf=RandomForestRegressor(n_estimators=20,max_depth=6,random_state=12)
model=rf.fit(xtrain,ytrain)
pred=model.predict(xtest)
pd.DataFrame(zip(pred,ytest))
mean_absolute_error(ytest,pred)
```

```python
from sklearn.model_selection import GridSearchCV
tp = {"n_estimators":range(2,50,1),"random_state":range(10,100,10),"max_depth":range(2,10,1)}
# tp = {"max_depth":range(2,10,1)}

gv=GridSearchCV(rf,tp,scoring="neg_mean_squared_error",cv=4)

cvmodel = gv.fit(xtrain,ytrain)
cvmodel.best_params_
```
