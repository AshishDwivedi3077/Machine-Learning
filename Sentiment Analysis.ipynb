```python
import pandas as pd

```

```python
data=pd.read_csv(r"C:\Users\ashishkumd2\Desktop\SOurcs\ML\Datsets\StockMarket.csv",encoding="ISO-8859-1")
# len(data)
data.head(1)
```

```python

train=data[data["Date"] < "201412"]
test=data[data["Date"] > "201412"]
len(train)
```

```python

data=train.iloc[:,2:27]
data.replace("[^a-zA-z]"," ",regex=True,inplace=True)

list1=[i for i in range(25)]
new_index=[str(i) for i in list1]
data.columns=new_index
data.head(1)
```

```python
for index in new_index:
    data[index]=data[index].str.lower()
data.head(5)    
```

```python
import re
headlines=[]
for row in range(0,len(data.index)):
    headlines.append(" ".join(str(x) for x in data.iloc[row,0:25]))

headlines[1]
```

```python
clean_headlines=[]
for h in range(0,len(headlines)):
    clean_headlines.append(re.sub("b",'',headlines[h]))
 
```

```python
clean_headlines[1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
CntVec=CountVectorizer(ngram_range=(3,3))
train_ds=CntVec.fit_transform(clean_headlines)
```

```python
train_ds[0]
```

```python
rf=RandomForestClassifier(n_estimators=150,criterion="entropy")
rf.fit(train_ds,train['Label'])
```

```python
test_tranforn=[]
for row in range(0,len(test.index)):
    test_tranforn.append(" ".join(str(x) for x in test.iloc[row,2:27]))
test_dataset=CntVec.transform(test_tranforn)   
pred=rf.predict(test_dataset)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
matrix=confusion_matrix(test["Label"],pred)
print(matrix)
score=accuracy_score(test["Label"],pred)
print(score)
cs=classification_report(test["Label"],pred)
print(cs)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k=KNeighborsClassifier(n_neighbors=1)
k.fit(train_ds,train['Label'])
pred = k.predict(test_dataset)
matrix=confusion_matrix(test["Label"],pred)
print(matrix)
score=accuracy_score(test["Label"],pred)
print(score)
cs=classification_report(test["Label"],pred)
print(cs)
```
