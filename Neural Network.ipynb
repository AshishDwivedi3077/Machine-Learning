```python
import pandas as pd
data=pd.read_csv(r"C:\Users\ashishkumd2\Desktop\SOurcs\ML\Datsets\Automobile_data.csv")
data.head(13)
```

```python
data['price'].replace('?',0,inplace=True)
```

```python
data.drop(['normalized-losses'], axis=1)
```

```python
data['fuel-type'].unique()
```

```python
X=data.drop(['symboling','aspiration','num-of-doors','drive-wheels','engine-location','body-style','normalized-losses','fuel-system','fuel-type'],axis=1)
X
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
X['make']=le.fit_transform(X['make'])#,'engine-type','num-of-cylinders'])
X['engine-type']=le.fit_transform(X['engine-type'])
X['num-of-cylinders']=le.fit_transform(X['num-of-cylinders'])
```

```python
X
```

```python
Y=data['fuel-type']
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.1,random_state=2)
```

```python
#!pip install keras 
xtrain.replace('?',0,inplace=True)#head(50)
```

```python

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()

model = lr.fit(xtrain,ytrain)
pred = model.predict(xtest)
```

```python
pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest,pred)
```
